```python
# Example code for bivariate pdfs
from scipy.stats import multivariate_normal as mn
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def plot_2d_Gaussian_pdf(means, cov):
    n = 100
    x1 = np.linspace(means[0] - 3 * np.sqrt(cov[0][0]), means[0] + 3 * np.sqrt(cov[0][0]), n)
    x2 = np.linspace(means[1] - 3 * np.sqrt(cov[1][1]), means[1] + 3 * np.sqrt(cov[1][1]), n)
    x1_v, x2_v = np.meshgrid(x1, x2)
    Xgrid = np.vstack([x1_v.ravel(), x2_v.ravel()]).T
    Y = mn.pdf(Xgrid, means, cov)
    fig, ax = plt.subplots()
    ax.pcolorfast(x1, x2, Y.reshape(x1_v.shape), alpha=0.5, cmap='Blues')
    ax.contour(x1_v, x2_v, Y.reshape(x1_v.shape),
               levels=[0.05, 0.1, 0.15, 0.2], alpha=0.3, colors='b')
    ax.axis('equal')
    ax.grid(alpha=0.2)
    plt.show()
    
means = [3, 2]
cov = [[1, 0.5], [0.5, 0.8]]
plot_2d_Gaussian_pdf(means, cov)
```

Using confidence intervals determine features for which you can state with confidence of 95% that the healthy population mean is larger/smaller than the disease population mean

```python
healthy = df[df["selector"] == 2]
sick = df[df["selector"] == 1]
def find_means(df1, df2, label):
    return df1[label].mean(), df2[label].mean()

def find_stds(df1, df2, label):
    return df1[label].std(ddof=1), df2[label].std(ddof=1)

def find_sample_error(df1, df2, label):
    std1, std2 = find_stds(df1, df2, label)
    return np.sqrt((std1 ** 2 / len(df1)) + (std2 ** 2 / len(df2)))

def calc_mean_interval(df1, df2, label, alpha):
    mean1, mean2 = find_means(df1, df2, label)
    sample_error = find_sample_error(df1, df2, label)
    n = len(df1) + len(df2)
    diff = mean1 - mean2
    dist = stats.t.ppf(1-alpha/2, n-2)
    return ((diff - dist) * sample_error, (diff + dist) * sample_error)

print('With 95% confidence, the following features means are larger/smaller')
for feature in 'age total_bili direct_bili alkphos sgpt sgot protein albumin ag_ratio'.split():
    bottom, top = calc_mean_interval(healthy, sick, feature, 0.05)
    mean1, mean2 = find_means(healthy, sick, feature)
    if bottom <= mean1 - mean2 <= top:
        print(f'{feature.ljust(20)}: \t\t{bottom:0.3f} <= {mean1-mean2:0.3f} <= {top:0.3f}')
```

```python
# Note: on Piazza we were told that the correlations had to be both high and significant
def show_correlations(df: pd.DataFrame):
    correlations = [stats.pearsonr, stats.spearmanr, stats.kendalltau]
    columns = 'age total_bili direct_bili alkphos sgpt sgot protein albumin ag_ratio'.split()
    for c1 in columns:
        for c2 in columns[columns.index(c1):]:
            if c1 == c2:
                continue
            flag = False
            for corr in correlations:
                rho, prob = corr(df[c1], df[c2])
                if abs(rho) >= 0.5 and prob < 0.05:
                    flag = True
                    print(f'{corr.__name__}\t{c1.ljust(11)}\t{c2.ljust(11)}\t{rho:0.3f}, {prob:0.3f}')
            if flag:
                flag = False
                print('=' * 10)
show_correlations(df)
```

### Question 4 - CLT for Markov chains (20 points)

In this exercise you will construct trajectories of Markovian dice rolling results in the following way.<br>
The first roll, X0, is Unif(1..6)<br>
After i rolls are determined the i+1st, Xi+1, is drawn according to the row that corresponds to the value of Xi in the matrix T below. <br>
In other words, T is the transition matrix of a Markov chain and the initial distribution is uniform.

\begin{equation*}
T = 
\begin{pmatrix}
0.5 & 0.2 & 0.05 & 0 & 0.05 & 0.2 \\
0.2 & 0.5 & 0.2 & 0.05 & 0 & 0.05 \\
0.05 & 0.2 & 0.5 & 0.2 & 0.05 & 0 \\
0 & 0.05 & 0.2 & 0.5 & 0.2 & 0.05 \\
0.05 & 0 & 0.05 & 0.2 & 0.5 & 0.2 \\
0.2 & 0.05 & 0 & 0.05 & 0.2 & 0.5
\end{pmatrix}
\end{equation*}

#### 4.A
Construct 1000 trajectories, each of length 20.
1. What do you expect the average value of all 20 numbers in a trajectory to be?
2. Compute the average value of each such trajectory. Draw a histogram of the 1000 numbers you received, using 100 bins.
3. What does the distribution look like? What are the empirical mean and the std?

1. We execpt the average value of all 20 number in a trajectory to be around 3.5. This is because the initial value is a fair dice, and the transition matrix seems to prefer we "cluster" around the main-diagonal. That is, once we get a value we tend to stick to that value. Therefore, we expect 3.5, which is the expectation of a fair dice. Also, notice that $p\cdot T = p$ which validates our intuition.

```python
# Part 2
prob_dict = {
    1: [0.5, 0.2, 0.05, 0, 0.05, 0.2],
    2: [0.2, 0.5, 0.2, 0.05, 0, 0.05],
    3: [0.05, 0.2, 0.5, 0.2, 0.05, 0],
    4: [0, 0.05, 0.2, 0.5, 0.2, 0.05],
    5: [0.05, 0, 0.05, 0.2, 0.5, 0.2],
    6: [0.2, 0.05, 0, 0.05, 0.2, 0.5]
}
def trajectory(length: int = 20):
    choices = [1, 2, 3, 4, 5, 6]
    val = np.random.randint(low=1, high=6)
    vals = [val]
    for _ in range(length - 1):
        val = np.random.choice(choices, p=prob_dict[vals[-1]])
        vals.append(val)
    return vals

def average(lst):
    return sum(lst) / len(lst)

averages = [average(trajectory()) for _ in range(1_000)]
plt.hist(averages, bins=100)
plt.show()
```

The distribution looks roughly normal.

```python
print(f"Empirical mean: {np.mean(averages):0.4f}")
print(f"Empirical std:  {np.std(averages):0.4f}")
```
